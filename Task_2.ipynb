```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data_files = ['/content/Unemployment in India.csv', '/content/Unemployment_Rate_upto_11_2020.csv']
```

```python
combined_data = pd.DataFrame()
```

```python
for data_file in data_files:
    data = pd.read_csv(data_file)
```

```python
print(f'Data from {data_file}:')
print(data.head())
print(data.info())
```

```python
print("Null values in the dataset:")
print(data.isnull().sum())
data = data.dropna()
```

```python
combined_data = combined_data.append(data, ignore_index=True)
```

```python
sns.set_style('whitegrid')
plt.figure(figsize=(12, 6))
```

```python
plt.figure(figsize=(12, 6))
```

```python
sns.lineplot(data=combined_data, x=' Date', y=' Estimated Unemployment Rate (%)', marker='o')
plt.title('Combined Unemployment Rate Over Time')
```

```python
plt.xlabel(' Date')
```

```python
plt.ylabel(' Estimated Unemployment Rate (%)')
```

```python
plt.xticks(rotation=45)
plt.show()
```

```python
plt.xlabel(' Date')
plt.ylabel(' Estimated Unemployment Rate (%)')
plt.xticks(rotation=45)
plt.show()
```

```python
sns.set_style('whitegrid')
plt.figure(figsize=(12, 6))
sns.lineplot(data=combined_data, x=' Date', y=' Estimated Unemployment Rate (%)', marker='o')
plt.title('Combined Unemployment Rate Over Time')
plt.xlabel(' Date')
plt.ylabel('  Estimated Unemployment Rate (%)')
plt.xticks(rotation=45)
plt.show()
```
